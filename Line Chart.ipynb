```python
import pandas as pd
import matplotlib.pyplot as plt
marathon_2015_2017 = pd.read_csv('./material/marathon_2015_2017.csv')
```

```python
record =  pd.DataFrame(marathon_2015_2017, columns=['5K', '10K', '15K', '20K', 'Half', '25K', '30K',
                                                    '35K', '40K', 'Official Time']).sort_values(by=[
    'Official Time'
])

record.insert(0, 'Rank', range(1, 1+len(record)))
record.head()
```

```python
top100 = record[0: 100]
top100
```

```python
xData = top100.Rank

yData_full = top100['Official Time']
yData_10K = top100['10K']
yData_20K = top100['20K']
yData_30K = top100['30K']

yData_full
```

```python
plt.figure(figsize=(20,10))

plt.plot(xData, yData_full)
plt.plot(xData, yData_10K)
plt.plot(xData, yData_20K)
plt.plot(xData, yData_30K)

plt.show()
```
